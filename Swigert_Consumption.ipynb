```python
state_row = []
child_description = []
for x in part_of_data_df.columns:
    split_list = x.split(", ")
    state_row.append([i for i in split_list if i in states_list][0])
    child_description.append(split_list[0])
```

```python
index_list = index_ls.append(Series(["state","child description"], index = [12,13]))
dummy_df = part_of_data_df.T
dummy_df["state"] = state_row
dummy_df["child description"] = child_description
#print dummy_df.head()
part_of_data_df = dummy_df.T
part_of_data_df
part_of_data_df = part_of_data_df.set_index(index_list)
part_of_data_df
```

```python
grouped_df = part_of_data_df.T.set_index(["child description","state","units"])

grouped_df = grouped_df.T.loc[["data"]]
grouped_df
```

```python
lista = []
data_ls = grouped_df.loc[["data"]].values
values_years_ls = DataFrame([i for i in zip(*data_ls[0][0])], index = ["year",grouped_df.columns[0]]).T.set_index("year")
lista.append(values_years_ls[grouped_df.columns[0]])

for x in range(1,len(data_ls[0])):
    data_rows = DataFrame([i for i in zip(*data_ls[0][x])], index = ["year",grouped_df.columns[x]]).T.set_index("year")
    lista.append(data_rows[grouped_df.columns[x]])

for x,ls in enumerate(lista):
    if x == 0:
        lista[x] = pd.DataFrame(ls)
        tablita = pd.DataFrame(lista[x])
    else:
        lista[x] = pd.DataFrame(ls)
        tablita = pd.merge(tablita,lista[x], how='outer', left_index=True, right_index=True)#, on="year")

tablita.columns = grouped_df.columns
organized_df = grouped_df.append(tablita)
organized_df["index"] = organized_df.index
organized_df = organized_df.drop_duplicates(cols='index', take_last=True, inplace=False)
organized_df.columns
one = organized_df.xs('index', level=0, axis=1)
del(one)
organized_df
```

```python
## A list of the child descriptions
children_list = organized_df.columns.levels[0].tolist()
children_list[0:6]
```

```python
## A list of the child descriptions
# includes UNITED STATES, WASHINGTON DC and an empty object
area_list = organized_df.columns.levels[1].tolist()
len(area_list)
```

```python
## We can select all available states for a specific child description
per_child = organized_df.xs((children_list[1]), level=("child description"), axis = 1)
per_child.head()
```

```python
## We can select for a specific STATE, or US, all its children:
per_area = organized_df.xs(area_list[0], level=("state"), axis = 1)
per_area.head()
```

```python
'''Selecting by both CHILD DESCRIPTION and STATE.  
This may be necessary for the cases where there are more than one column because of the units '''

per_child_and_area = organized_df.xs((children_list[1], area_list[1]), level=("child description","state"), axis = 1)
per_child_and_area
```
