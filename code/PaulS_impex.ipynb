```python
grouped_df = part_of_data_df.T.set_index(["child description","state","units"])

grouped_df = grouped_df.T.loc[["data"]]
grouped_df
```

```python
lista = []
data_ls = grouped_df.loc[["data"]].values
values_years_ls = DataFrame([i for i in zip(*data_ls[0][0])], index = ["year",grouped_df.columns[0]]).T.set_index("year")
lista.append(values_years_ls[grouped_df.columns[0]])

for x in range(1,len(data_ls[0])):
    data_rows = DataFrame([i for i in zip(*data_ls[0][x])], index = ["year",grouped_df.columns[x]]).T.set_index("year")
    lista.append(data_rows[grouped_df.columns[x]])

for x,ls in enumerate(lista):
    if x == 0:
        lista[x] = pd.DataFrame(ls)
        tablita = pd.DataFrame(lista[x])
    else:
        lista[x] = pd.DataFrame(ls)
        tablita = pd.merge(tablita,lista[x], how='outer', left_index=True, right_index=True)#, on="year")

tablita.columns = grouped_df.columns
organized_df = grouped_df.append(tablita)
organized_df["index"] = organized_df.index
organized_df = organized_df.drop_duplicates(cols='index', take_last=True, inplace=False)
organized_df.columns
one = organized_df.xs('index', level=0, axis=1)
del(one)
organized_df
```
