In this section, we read in annual state-by-state energy production (also called generation) data from the ELEC.txt file. The units of this data set are thousands of megawatt hours. 

For net generation by source, we want to read in all JSON lines of the ELEC.txt file if they match the format: 'Net Generation: SOURCE NAME : STATE NAME : All Sectors : Annual'

```python
#set up regex to extract JSON lines of interest
data_match = re.compile('Net Generation : .*?: .*?: All Sectors : Annual')

raw_data = []
with open(e_location) as myfile:
    for row in myfile:
        line = json.loads(row)
        if data_match.search(line['name']):
            raw_data.append(line)
```

We can put everything into a dictionary first, then turn it into a data frame.

```python
#Create dictionary from the filtered_data json

#set up regular expressions to match the state and source
state_match = re.compile('Net Generation : .*? : (.*?) : All Sectors : Annual')
source_match = re.compile('Net Generation : (.*?) : .*? : All Sectors : Annual')

#create dictionary to store results

total_dict = defaultdict(dict)

#loop through filtered_data to populate data_dict
for x in filtered_data:
    year_dict = {}
    source_dict = {}
    data = x['data']
    name = x['name']
    state = state_match.findall(name)[0]
    source = source_match.findall(name)[0]
    #change the source to all lowercase,  _ instead of spaces, and no parentheses
    source = source.replace(" ","_").replace("(","").replace(")","").lower()
    for pair in data:
        year_dict[pair[0]] = pair[1]
    total_dict[state][source] = year_dict
total_dict
```

```python
#turn the nested dictionary into a pandas DataFrame
#see helpful reference at http://stackoverflow.com/questions/5900578/how-collections-defaultdict-work

states = []
sources = []

for state, source_dict in total_dict.iteritems():
    states.append(state)
    sources.append(pd.DataFrame.from_dict(source_dict))
    
source_frame = pd.concat(sources, keys=states)
```

This section obtains data on energy price and revenue by sectors from the ELEC.txt file. The units of price are in cents per kilowatt hour. The units for revenue are in millions of dollars. 

### Read in data

```python
ELEC_data = []
retail_sales = []
revenue_retail_sales = []
avg_elec_price = []
avg_cost_fossil = []
with open(e_location, 'r') as f:
    for line in f: 
        s = json.loads(line)   
        if any(x in s['name'] for x in ["Price","Cost","Retail","Sales","Revenue"]):#
            if "Annual" in s['name']:
                ELEC_data.append(s)
        if "Retail Sales of Electricity" in s['name'][0:27]:
            if "Annual" in s['name']:
                retail_sales.append(s)
        if "Revenue from Retail Sales of Electricity" in s['name']:
            if "Annual" in s['name']:
                revenue_retail_sales.append(s)
        if "Average Retail Price of Electricity" in s['name']:
            if "Annual" in s['name']:
                avg_elec_price.append(s)        
        if "Average Cost of Fossil Fuels (per BTU)" in s['name']:
            if "Annual" in s['name']:
                avg_cost_fossil.append(s)       
```

## Imports and Exports

This section reads in data on annual state-by-state energy imports and exports from the SEDS.txt file. The units of both imports and exports are in billions of BTU. For consistency with the production and sales data, we later convert this to gigawatt hours. 

### Read in Data

```python
# Accessing the bulk data file to get Import and Export Electricity
def get_ie(location):
    data_match = re.compile('Electricity .*? the United States, .*?')
    units_match = re.compile('Billion Btu')
    raw_data = []
    with open(location) as myfile:
        for row in myfile:
            line = json.loads(row)
            if (data_match.search(line['name'])) and (units_match.search(line['units'])):
                raw_data.append(line)
        return raw_data

raw_data = get_ie(s_location)
len(raw_data)
```

```python
# create fcn to filter out any 'data' before 2001 to match with the other datasets

def filter_data(raw_data):
    the_range = len(raw_data)
    for number in range(the_range):
        int_numb = int(number)
        filtered_data = [x for x in raw_data[int_numb]['data'] if int(x[0]) > 2000]
        raw_data[int_numb]['data'] = filtered_data
    return raw_data
```

```python
#Create dictionary from the filtered_data json

#set up regular expressions to match the state and source
state_match = re.compile('Net Generation : .*? : (.*?) : All Sectors : Annual')
source_match = re.compile('Net Generation : (.*?) : .*? : All Sectors : Annual')

#create dictionary to store results

total_dict = defaultdict(dict)

#loop through filtered_data to populate data_dict
def make_dict(json_data):
    for x in json_data:
        year_dict = {}
        source_dict = {}
        #The list of lists
        data = x['data']
        # The name
        name = x['name']
        state = name[45:]
        source = name[:18]
        #change the source to all lowercase,  _ instead of spaces, and no parentheses
        source = source.lower()
        for pair in data:
            year_dict[pair[0]] = pair[1]
        total_dict[state][source] = year_dict
    return total_dict


total_dict = make_dict(filtered_data)
total_dict['Texas']
```

```python
#turn the nested dictionary into a pandas DataFrame
#see helpful reference at http://stackoverflow.com/questions/5900578/how-collections-defaultdict-work

states = []
sources = []

for state, source_dict in total_dict.iteritems():
    states.append(state)
    sources.append(pd.DataFrame.from_dict(source_dict))
    
source_frame = pd.concat(sources, keys=states)
```

## Summarization

In this section, we aggregate and summarize the data to measure the percentage different of each state's variables from the US state average for that year. 

```python
all_data.head()
```

```python
#get the state average for each variable and year
state_average_by_year = all_data.groupby(level=1).apply(np.mean)
state_average_by_year.head()
```

```python
#calculate each state's percentage difference from the US average for each year. 
all_percentage_data = []

for state in state_names:
    by_state = all_data.xs(state,level = 'state', axis=0)
    #US_mean = 
    percentage_by_state = (by_state - state_average_by_year)/state_average_by_year*100
    percentage_by_state['state']= state
    percentage_by_state['year']=percentage_by_state.index
    percentage_by_state = percentage_by_state.set_index(['state','year'])       
    all_percentage_data.append(percentage_by_state)
    
all_perc_table=pd.concat(all_percentage_data, axis = 0)
all_perc_table[0:13]
```

```python
all_perc_table.columns
```

Now we do the same thing but get the percentage difference summed across all years

```python
#get the state average for each variable and year
state_totals = all_data.groupby(level=0).apply(np.sum)
state_mean_total_all_years = state_totals.apply(np.mean)
state_mean_total_all_years.head()
```

```python
#calculate each state's percentage difference from the US average for each year. 
all_percentage_sum_data = []

for state in state_names:
    by_state = state_totals.xs(state,axis=0)
    #US_mean = 
    percentage_by_state = (by_state - state_mean_total_all_years)/state_mean_total_all_years*100
    percentage_by_state = DataFrame(percentage_by_state)
    percentage_by_state = percentage_by_state.transpose()
    percentage_by_state['state']= state
    percentage_by_state.set_index('state', inplace=True)       
    all_percentage_sum_data.append(percentage_by_state)
    
all_perc_sum_table=pd.concat(all_percentage_sum_data, axis = 0)
all_perc_sum_table[0:13]
```
