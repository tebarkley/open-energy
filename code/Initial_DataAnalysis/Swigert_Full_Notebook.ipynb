### Okay let's get the state populations now!!

```python
import census
import settings
import us
c=census.Census(settings.CENSUS_KEY)

# call the API and instantiate `df`
c_pop_df = DataFrame(c.sf1.get('NAME,P0010001', geo={'for':'state:*'}))
# convert the population to integer
c_pop_df['P0010001'] = c_pop_df['P0010001'].astype(np.int)
c_pop_df.rename(columns={'P0010001':'Population','state':'FIPS'}, inplace=True) 
#Get rid of Puerto Rico
c_pop_df = c_pop_df[c_pop_df['NAME'] != 'Puerto Rico']
print c_pop_df
```

```python
#Now I'm going to set the state names as indices for both of the tables (energy consumption and population)
c_pop_df = c_pop_df.set_index(['NAME'])
c_pop_df.head()
```

```python
#Let's calculate the per capita used now
energy_pop_df = just_states_df.join(c_pop_df)
energy_pop_df['p_All_Sectors'] = (energy_pop_df['All Sectors'].astype(float) / (energy_pop_df['Population']).astype(float))
energy_pop_df['p_Commercial'] = (energy_pop_df['Commercial'].astype(float) / (energy_pop_df['Population']).astype(float))
energy_pop_df['p_Industrial'] = (energy_pop_df['Industrial'].astype(float) / (energy_pop_df['Population']).astype(float))
energy_pop_df['p_Residential'] = (energy_pop_df['Residential'].astype(float) / (energy_pop_df['Population']).astype(float))

energy_pop_df.head()
```

```python
energy_pop_df_perc = energy_pop_df[['p_All_Sectors', 'p_Commercial', 'p_Industrial', 'p_Residential']]
energy_pop_df_perc.to_csv("with_percentages")
```
