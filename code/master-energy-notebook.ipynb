## Net Generation by Source

For net generation by source, we want to read in all JSON lines of the ELEC.txt file if they match the format: 'Net Generation: SOURCE NAME : STATE NAME : All Sectors : Annual'

```python
#set up regex to extract JSON lines of interest
data_match = re.compile('Net Generation : .*?: .*?: All Sectors : Annual')

raw_data = []
with open(e_location) as myfile:
    for row in myfile:
        line = json.loads(row)
        if data_match.search(line['name']):
            raw_data.append(line)
```

We can look at a list of all of the 'name' JSON elements. We can see that there are some entries that are not for states (like New England, Pacific Contiguous, etc)

```python
names = []
for x in raw_data:
    names.append(x['name'])
names
```

We can use the US module to obtain a list of US states, then filter the raw_data list down to only entries for a US state

```python
#make a list of all of the US state names- we want to get rid of all data rows that are not from individual states
state_names = [state.name for state in states.STATES]
```

```python
#set up regex to extract the state name
state_match = re.compile('Net Generation : .*? : (.*?) : All Sectors : Annual')

#create filtered_data list that only contains state data
filtered_data = []
for row in raw_data:
    name = row['name']
    if state_match.findall(name)[0] in state_names:
        filtered_data.append(row)
```

From the above, we can see that the year is element 0 of each list within the data list, and the data value is element 1. The state is located in a predictable place within the name string. The type of data is also located in a predictable place within the name string.

We can put everything into a dictionary first, then turn it into a data frame.

```python
#Create dictionary from the filtered_data json

#set up regular expressions to match the state and source
state_match = re.compile('Net Generation : .*? : (.*?) : All Sectors : Annual')
source_match = re.compile('Net Generation : (.*?) : .*? : All Sectors : Annual')

#create dictionary to store results

total_dict = defaultdict(dict)

#loop through filtered_data to populate data_dict
for x in filtered_data:
    year_dict = {}
    source_dict = {}
    data = x['data']
    name = x['name']
    state = state_match.findall(name)[0]
    source = source_match.findall(name)[0]
    #change the source to all lowercase,  _ instead of spaces, and no parentheses
    source = source.replace(" ","_").replace("(","").replace(")","").lower()
    for pair in data:
        year_dict[pair[0]] = pair[1]
    total_dict[state][source] = year_dict
total_dict
```

```python
#turn the nested dictionary into a pandas DataFrame
#see helpful reference at http://stackoverflow.com/questions/5900578/how-collections-defaultdict-work

states = []
sources = []

for state, source_dict in total_dict.iteritems():
    states.append(state)
    sources.append(pd.DataFrame.from_dict(source_dict))
    
source_frame = pd.concat(sources, keys=states)
```

#### Preparing the dataframes

```python
avg_elec_price_name = []
retail_sales_name = []
revenue_retail_sales_name = []

for retail_name, price_name, rev_name in map(None, retail_sales, avg_elec_price, revenue_retail_sales):
    avg_elec_price_name.append(price_name['name'])
    retail_sales_name.append(retail_name['name'])
    revenue_retail_sales_name.append(rev_name['name'])
```

```python
ELEC_data_df = []
retail_sales_df = []
revenue_retail_sales_df = []
avg_elec_price_df = []
avg_cost_fossil_df = []

for price, sales, revenue in map(None,avg_elec_price,retail_sales, revenue_retail_sales):
    index_ls = pd.DataFrame(price.keys())
    retail_sales_df.append(pd.DataFrame(sales.values()))
    avg_elec_price_df.append(pd.DataFrame(price.values()))
    revenue_retail_sales_df.append(pd.DataFrame(revenue.values()))
    
avg_elec_price_df = pd.concat(avg_elec_price_df, axis=1, keys = avg_elec_price_name).xs(0, level=1, axis = 1)
avg_elec_price_df['index'] = index_ls
avg_elec_price_df.set_index('index', drop=True,inplace = True)

retail_sales_df = pd.concat(retail_sales_df, axis=1, keys = retail_sales_name).xs(0, level=1, axis = 1)
retail_sales_df['index'] = index_ls
retail_sales_df.set_index('index', drop=True,inplace = True)

revenue_retail_sales_df = pd.concat(revenue_retail_sales_df, axis=1, keys = revenue_retail_sales_name).xs(0, level=1, axis = 1)
revenue_retail_sales_df['index'] = index_ls
revenue_retail_sales_df.set_index('index', drop=True,inplace = True)
```

```python
area_p_ls, area_s_ls, area_r_ls = [], [], []
sector_p_ls, sector_s_ls, sector_r_ls = [], [], []

for price, sales, revenue in map(None,avg_elec_price_df,retail_sales_df, revenue_retail_sales_df):
    price_split, sales_split, revenue_split = price.split(" : "), sales.split(" : "), revenue.split(" : ")
    area_p_ls.append(price_split[1]), area_s_ls.append(sales_split[1]), area_r_ls.append(revenue_split[1])
    sector_p_ls.append(price_split[2]), sector_s_ls.append(sales_split[2]), sector_r_ls.append(revenue_split[2])
    
dummy_df_p, dummy_df_s, dummy_df_r = avg_elec_price_df.T, retail_sales_df.T, revenue_retail_sales_df.T
dummy_df_p["region"], dummy_df_s["region"], dummy_df_r["region"] = area_p_ls, area_s_ls, area_r_ls
dummy_df_p["sector"], dummy_df_s["sector"], dummy_df_r["sector"] = sector_p_ls, sector_s_ls, sector_r_ls

avg_elec_price_df,retail_sales_df, revenue_retail_sales_df = dummy_df_p.T, dummy_df_s.T, dummy_df_r.T

avg_elec_price_df.columns.name,retail_sales_df.columns.name, revenue_retail_sales_df.columns.name = price_split[0], sales_split[0], revenue_split[0]
```

```python
grouped_by_cat = ['sector','region']

grouped_price_df = avg_elec_price_df.T.set_index(grouped_by_cat)
grouped_sales_df = retail_sales_df.T.set_index(grouped_by_cat)
grouped_revenue_df = revenue_retail_sales_df.T.set_index(grouped_by_cat)

grouped_price_df = grouped_price_df.T.loc[["data"]]
grouped_sales_df = grouped_sales_df.T.loc[["data"]]
grouped_revenue_df = grouped_revenue_df.T.loc[["data"]]

grouped_price_df.index.name=avg_elec_price_df.columns.name
grouped_sales_df.index.name=retail_sales_df.columns.name
grouped_revenue_df.index.name=revenue_retail_sales_df.columns.name
```

```python
unicode_P_ls, unicode_S_ls, unicode_R_ls = [], [], []

## PRICES
for x , title in enumerate(grouped_price_df.columns):
    if type(grouped_price_df.iloc[0,x]) == unicode:
        unicode_P_ls.append([x,title])

for x, uni in enumerate(unicode_P_ls):
    grouped_price_df = grouped_price_df.drop(uni[1],1)

# SALES    
for x , title in enumerate(grouped_sales_df.columns):
    if type(grouped_sales_df.iloc[0,x]) == unicode:
        unicode_S_ls.append([x,title])

for x, uni in enumerate(unicode_S_ls):
    grouped_sales_df = grouped_sales_df.drop(uni[1],1)
    
# REVENUE    
for x , title in enumerate(grouped_revenue_df.columns):
    if type(grouped_revenue_df.iloc[0,x]) == unicode:
        unicode_R_ls.append([x,title])

for x, uni in enumerate(unicode_R_ls):
    grouped_revenue_df = grouped_revenue_df.drop(uni[1],1)
```

```python
grouped_ls = [grouped_price_df, grouped_sales_df, grouped_revenue_df]

prepared_name_ls = ['Price_df', 'Sales_df', 'Revenue_df']
cat_ls = ['Price', 'Sales', 'Revenue']
prepared_results = {}

for x, ls in enumerate(grouped_ls):
    lista = []
    data_ls = ls.loc[["data"]].values
    values_years_ls = DataFrame([i for i in zip(*data_ls[0][0])], index = ["year",ls.columns[0]]).T.set_index("year")
    lista.append(values_years_ls[ls.columns[0]])

    for y in range(1,len(data_ls[0])):
        data_rows = DataFrame([i for i in zip(*data_ls[0][y])], index = ["year",ls.columns[y]]).T.set_index("year")
        lista.append(data_rows[ls.columns[y]])

    for z in range(0,len(lista)):
        if z == 0:
            tablita = pd.DataFrame(lista[z])
        else:
            tutu = pd.DataFrame(lista[z])
            tablita = pd.merge(tablita,tutu, how='outer', left_index=True, right_index=True)#, on="year")
    #print "HOLA"
    tablita.columns = ls.columns
    organized_df = ls.append(tablita)
    organized_df["index"] = organized_df.index
    organized_df = organized_df.drop_duplicates(cols='index', take_last=True, inplace=False)
    one = organized_df.xs('index', level=0, axis=1)
    del(one)
    organized_df = organized_df.drop(["data"])
    
    # Add level with category name
    org_T = organized_df.T
    org_T['category'] = cat_ls[x]
    org_T.set_index("category", append=True, inplace=True)
    org_T = org_T.reorder_levels(['category','sector','region'])#.T
    complete_df = org_T.T
    
    # Add dataframe to the dictionary
    prepared_results["%s" % prepared_name_ls[x]] = complete_df
    del organized_df
    del tablita
    del complete_df
```

```python
''' And voila! This is the yearly result after processing ELEC bulk file. '''

complete_df = pd.concat([prepared_results["Price_df"],prepared_results["Sales_df"],prepared_results["Revenue_df"]], axis=1).sort_index()
complete_df.head()
```

```python
# Accessing the bulk data file to get Import and Export Electricity
def get_ie(location):
    #e_location = '/Users/paulson/Desktop/MIMS/i290T/bulkdata/SEDS.txt'
    data_match = re.compile('Electricity .*? the United States, .*?')
    units_match = re.compile('Billion Btu')
    raw_data = []
    with open(location) as myfile:
        for row in myfile:
            line = json.loads(row)
            if (data_match.search(line['name'])) and (units_match.search(line['units'])):
                raw_data.append(line)
        return raw_data

raw_data = get_ie(s_location)
len(raw_data)
```

```python
# create fcn to filter out any 'data' before 2001

def filter_data(raw_data):
    the_range = len(raw_data)
    for number in range(the_range):
        int_numb = int(number)
        filtered_data = [x for x in raw_data[int_numb]['data'] if int(x[0]) > 2000]
        raw_data[int_numb]['data'] = filtered_data
    return raw_data
```
