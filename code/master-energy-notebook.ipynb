## Net Generation by Source

For net generation by source, we want to read in all JSON lines of the ELEC.txt file if they match the format: 'Net Generation: SOURCE NAME : STATE NAME : All Sectors : Annual'

```python
#set up regex to extract JSON lines of interest
data_match = re.compile('Net Generation : .*?: .*?: All Sectors : Annual')

raw_data = []
with open(e_location) as myfile:
    for row in myfile:
        line = json.loads(row)
        if data_match.search(line['name']):
            raw_data.append(line)
```

We can look at a list of all of the 'name' JSON elements. We can see that there are some entries that are not for states (like New England, Pacific Contiguous, etc)

```python
names = []
for x in raw_data:
    names.append(x['name'])
names
```

We can use the US module to obtain a list of US states, then filter the raw_data list down to only entries for a US state

```python
#make a list of all of the US state names- we want to get rid of all data rows that are not from individual states
state_names = [state.name for state in states.STATES]
```

```python
#set up regex to extract the state name
state_match = re.compile('Net Generation : .*? : (.*?) : All Sectors : Annual')

#create filtered_data list that only contains state data
filtered_data = []
for row in raw_data:
    name = row['name']
    if state_match.findall(name)[0] in state_names:
        filtered_data.append(row)
```

```python
len(raw_data) - len(filtered_data)
```

```python
names = []
for x in filtered_data:
    names.append(x['name'])
names
```

Now, we can create a dictionary of SOURCE_NAME : Description so that we can interpret what the different sources are

```python

#set up regex to extract the unique parts of the name
name_match = re.compile('Net Generation : (.*?) : .*? : All Sectors : Annual')

#set up dictionary in which to store results
descriptions = defaultdict(str)

for x in filtered_data:
    name = x['name']
    unique_name = name_match.findall(x['name'])[0]
    descriptions[unique_name] = x['description']

descriptions
```

Now, we can put the data into a dataframe with the state and year as the row indices, and the source as the column index

```python
#take a look at the form of the data
filtered_data[0]
```

From the above, we can see that the year is element 0 of each list within the data list, and the data value is element 1. The state is located in a predictable place within the name string. The type of data is also located in a predictable place within the name string.

We can put everything into a dictionary first, then turn it into a data frame.

```python
#Create dictionary from the filtered_data json

#set up regular expressions to match the state and source
state_match = re.compile('Net Generation : .*? : (.*?) : All Sectors : Annual')
source_match = re.compile('Net Generation : (.*?) : .*? : All Sectors : Annual')

#create dictionary to store results

total_dict = defaultdict(dict)

#loop through filtered_data to populate data_dict
for x in filtered_data:
    year_dict = {}
    source_dict = {}
    data = x['data']
    name = x['name']
    state = state_match.findall(name)[0]
    source = source_match.findall(name)[0]
    #change the source to all lowercase,  _ instead of spaces, and no parentheses
    source = source.replace(" ","_").replace("(","").replace(")","").lower()
    for pair in data:
        year_dict[pair[0]] = pair[1]
    total_dict[state][source] = year_dict
total_dict
```

```python
#turn the nested dictionary into a pandas DataFrame
#see helpful reference at http://stackoverflow.com/questions/5900578/how-collections-defaultdict-work

states = []
sources = []

for state, source_dict in total_dict.iteritems():
    states.append(state)
    sources.append(pd.DataFrame.from_dict(source_dict))
    
source_frame = pd.concat(sources, keys=states)
```
